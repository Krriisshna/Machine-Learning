# Vectors

## Vector Creation

```python
# NumPy routines which allocate memory and fill arrays with value
a = np.zeros(4)
# here argument is integer
print(f"np.zeros(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.zeros((4,))
# here argument is tuple with only one value(4)
print(f"np.zeros(4,): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.random_sample(4)
# default value 0.0≤x<1.0
print(f"np.random.random_sample(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

```python
# NumPy routines which allocate memory and fill arrays with value but do not accept shape as input argument
a = np.arange(4.)
print(f"np.arange(4.): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.rand(3)
print(f"np.random.rand(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
# random_sample() and rand() do the same thing
# difference is their argument differ as random_sample() can take a tuple
```

```python
# NumPy routines which allocate memory and fill with user specified values
a = np.array([5,4,3,2]);  print(f"np.array([5,4,3,2]):  a = {a},     a shape = {a.shape}, a data type = {a.dtype}")
a = np.array([5.,4,3,2]); print(f"np.array([5.,4,3,2]): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

## Operations on Vectors

### Indexing

```python
a = np.arange(10)
print(a)

# access an element
print(f"a[2].shape = {a[2].shape} a[2] = {a[2]}, Accessing an element returns a scalar")

# negative indexes count from the end
print(f"a[-1] = {a[-1]}")

# indexes must be within the range of the vector or they will produce an error.
try:
    c = a[10]
except Exception as e:
    print("The error message you'll see is:")
    print(e)
```

### Slicing
- Slicing use set of three values [Start:Stop:Stop]

```python
a = np.arange(10)
print(f"a = {a}")

c = a[2:7:1]
print(c)
c = a[2:7:2]
print(c)
c = a[3:]
print(c)
c = a[:3]
print(c)
c = a[:]
print(c)
```

### Single Vector Operations

```python
a = np.array([1,2,3,4])
print(f"a = {a}")
# negate elements of a
b = -a
print(f"b = -a: {b}")

b = np.sum(a)
print(f"b = np.sum(a): {b}")

b = np.mean(a)
print(f"b = np.mean(a): {b}")

b = a**2
print(f"b = a**2: {b}")
```

### Vector Vector element-wise operations
- Most of the NumPy arithmetic, logical and comparison operatins apply to vectors as well.  
$c_i$= $a_i + b_i$


```python
a = np.array([1,2,3,4])
b = np.array([-1,-2,3,4])
print(f"Binary operator work element wise: {a + b}")
# Vector size must be same
```

```python
# try a mismatched vector size
c = np.array([1,2])
try:
    d = a + c
except Exception as e:
    print("The error message you'll see is:")
    print(e)
```

### Scalar Vector operations

```python
a = np.array([1,2,3,4])

b = 5*a
print(f"b = 5*a: {b}")
```

### Vector Vector dot product
- dot product returns a scalar value

```python
# using for loop
# assume a and b are of same size
def my_dot(a, b):
    x = 0
    for i in range(a.shape[0]):
        x = x + a[i] * b[i]

    return x
```

```python
# test 1-D
a = np.array([1,2,3,4])
b = np.array([-1,4,3,2])
print(f"my_dot(a,b) = {my_dot(a,b)}")
```

```python
# now using np.dot on 1-D
a = np.array([1,2,3,4])
b = np.array([-1,4,3,2])
c = np.dot(a,b)
print(f"NumPy 1-D np.dot(a,b) = {c}, np.dot(a,b).shape = {c.shape}")
c = np.dot(b,a)
print(f"NumPy 1-D np.dot(b,a) = {c}, np.dot(b,a).shape = {c.shape}")
```

### The Need for Speed: vector vs for loop

```python
np.random.seed(1)
# seed() is like random_state but here it is for producing arrays
a = np.random.rand(10000000)
b = np.random.rand(10000000)
tic = time.time() # tic capture start time
c = np.dot(a,b)
toc = time.time()# toc capture end time
# time.time() returns the total seconds passed since Jan 1, 1970 00:00:00 UTC
# up to the moment time.time() is executed
print(f"np.dot(a,b) = {c:.4f}")
print(f"Vectorized version duration: {1000*(toc-tic):.4f} ms")

tic = time.time()
c = my_dot(a,b)
toc = time.time()

print(f"my_dot(a,b) = {c:.4f}")
print(f"Loop version duration: {1000*(toc-tic):.4f} ms")

del(a)
del(b)
# remove these big arrays from memory
```
